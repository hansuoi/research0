- txtをマージするコード

```python
with open('a.txt', 'r', encoding='utf-8') as f:
    first  = f.readlines()
with open('b.txt', 'r', encoding='utf-8') as f:
    latter = f.readlines()

print('first  = {}'.format(first))
print('latter = {}'.format(latter))
```

```python
for file in [first, latter]:
    for i in range(len(file)):
        if file[i][len(file[i])-1] != '\n':
            file[i] += '\n'

print('first  = {}'.format(first))
print('latter = {}'.format(latter))
```

```python
with open('c.txt', 'w', encoding='utf-8') as f:
    f.writelines(first)
    f.writelines(latter)

with open('c.txt', 'r', encoding='utf-8') as f:
    result = f.readlines()
print('result = {}'.format(result))
```

- csvをマージするコード

```python
import csv

with open('a.csv', 'r', encoding='utf-8') as f1:
    with open('b.csv', 'r', encoding='utf-8') as f2:
        first  = csv.reader(f1, delimiter=',')
        latter = csv.reader(f2, delimiter=',')
        with open('c.csv', 'w', encoding='utf-8', newline='') as f3:
            writer = csv.writer(f3)
            for f in first:
                writer.writerow(f)
            for l in latter:
                writer.writerow(l)

with open('c.csv', 'r', encoding='utf-8') as f:
    result = csv.reader(f)
    print('result = ')
    for i in result:
        print(i)
```
