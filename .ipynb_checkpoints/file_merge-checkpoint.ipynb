- txtをマージするコード

```python
with open('a.txt', 'r', encoding='utf-8') as f:
    first  = f.readlines()
with open('b.txt', 'r', encoding='utf-8') as f:
    latter = f.readlines()

print('first  = {}'.format(first))
print('latter = {}'.format(latter))
```

```python
for file in [first, latter]:
    for i in range(len(file)):
        if file[i][len(file[i])-1] != '\n':
            file[i] += '\n'

print('first  = {}'.format(first))
print('latter = {}'.format(latter))
```

```python
with open('c.txt', 'w', encoding='utf-8') as f:
    f.writelines(first)
    f.writelines(latter)

with open('c.txt', 'r', encoding='utf-8') as f:
    result = f.readlines()
print('result = {}'.format(result))
```

- csvをマージするコード

```python
import csv

with open('a.csv', 'r', encoding='utf-8') as f1:
    with open('b.csv', 'r', encoding='utf-8') as f2:
        first  = csv.reader(f1)
        latter = csv.reader(f2)
        with open('c.csv', 'w', encoding='utf-8') as f3:
            writer = csv.writer(f3, lineterminator='\n')
            writer.writerow(f1)
            writer.writerow(f2)
```

```python
with open('../middle/.csv', 'r', encoding='utf-8') as f:
    first  = csv.reader(f)
    for i in first:
        print(i)
```
